```python
#J0403231081_Frizaski Al Fath

# Mengenal Function

#How to Call a Function
def fungsi():
    print("Welcome to my code")

fungsi()

#Parameters & Arguments

def penjumlahan(a,b):
    print(f"{a} + {b} = {a+b}")

penjumlahan(10,15)

#Arguments-Positional
def ucapan(nama,selamat):
    print(f"{selamat}, {nama}!")

ucapan("Alfath","Hai")

#Arguments-Keyword
def selamat(name,ucapan):
    print(f"{ucapan}, {name}")

selamat(ucapan="Selamat datang",name="Alfath")

#Arguments-Default
def hadiah(name,ucapan="Selamat kamu menang undian"):
    print(f"{ucapan}, {name}!")

hadiah("Alfath")

#Arguments-Arbitrary-*Argy
# * Agar bisa mencetak beberapa argumen
def list_argumen(*args):
    for argumen in args:
        print(argumen)

list_argumen(1,"apple",True)

#Arguments-Arbitrary-**Kwargs
# **kwargs agar bisa melangkahi nomor variabel argumen
def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")
# key = nama variabel, value = yang diinput
print_kwargs(name="Alfath", umur=18)

#Arguments-Passing Lists/Dictionaries/Etc.
def print_list(angka):
    for num in angka:
        print(num)

print_list([1,2,3,4],)

def print_kamus(my_kamus):
    for key,value in my_kamus.items():
        print(f"{key}:{value}")

print_kamus({"nama":"Alfath","Umur":18})

#Different argument combinations

def f(a,*b,c=6,**d):
    print(f"a:{a}")
    print(f"b:{b}")
    print(f"c:{c}")
    print(f"d:{d}")

f(1,2,3,c=7,x=4,y=5)

#compare with

def f(a,b=6,*c,**d):
    print(f"a:{a}")
    print(f"b:{b}")
    print(f"c:{c}")
    print(f"d:{d}")

f(1,2,3,c=7,x=4,y=5)

#RETURN STATEMENT
# def add_angka(angka_1,angka_2):
#     result = a + b
#     return result

# sum_result = add_angka(15,3)
# print(f"The sum is:{sum_result}")

if"a"=="a":
    def greet():
        return "halo penghuni bumi"
else:
    def greet():
        return "selamat pagi bro!"
greetings = greet


def say_hello():
    print("Hai dunia")

greeting = say_hello
greeting()

# def say_hello():
#     print("Hello, Python!")

# say_hello()

# #Executed at Runtime
# def divide_by_zero(a):
#     return a*2/(a-a) this is a bug

# print("after function creation")

# #it's not error until executed:
# result = divide_by_zero(5)
# print("after function is calles/executed")

#It is first class object
def greet(nama):
    return f"Hello,{f'Farhan'}"

#Assign the function to a variable
my_function = greet

#Call the function
result = my_function("Farhan")
print(result)

#It is first class object-another examples
#1
def apply(func,x):
    return func(x)

def rumus_kotak(x):
    return x*x

result = apply(rumus_kotak,6)
print(result)

#2
def get_multiplier(factor):
    def multiplier(x):
        return x*factor
    return multiplier

double = get_multiplier(2)
triple = get_multiplier(3)
quadra = get_multiplier(4)

print(double(5))
print(triple(5))
print(quadra(6))

#Scope
def my_function():
    x=10 #x ini adalah local scope
    print(x)

def outer_function():
    y=20 #y ini adalah enclosing scope
    def inner_function():
        print(y) #inner_function bisa mengakses y dari enclosing scope

z=30 #z ini adalah global scope
def another_function():
    print(z) #another_function bisa mengakses z dari global scope

print(len("Hai Dunia")) #len itu bawaan dari function

#Scope-global keyword
global_va=10 #Global variable

def modify_global():
    global global_var #menggunakan global keyword untuk modify global variable
    global_var=30

modify_global()
print(global_var) 

#pass by object reference
#1
def modify_list(my_list):
    my_list.append(10) #memodifikasi list di dalam function, menambahkan angka 10 diakhir

my_list=[1,2,3]
modify_list(my_list)
print(my_list)

def reassign_list(my_list):
    my_list = [4,5,6] #reassigning list ke objek baru

my_list = [1,2,3]
reassign_list(my_list)
print(my_list)

#2
def modify_integer(x):
    x += 1 #memodifikasi integer didalam function

my_integer = 18
modify_integer(my_integer)
print(my_integer)

#Function overloading
def product(a,b):
    print(a*b)

def product(a,b,c):
    print(a*b*c)

#product(4,5) # kalau komentarnya dihilangkan maka fungsi akan error
product(4,3,5)

#Function Recursion
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

result = factorial(6)
print(result)

#Error Handling
def divide(x,y):
    try:
        result=x/y
    except ZeroDivisionError:
        print("Error:Dvision by zero")
        return None
    except TypeError:
        print("Error:Invalid data types")
        return None
    except Exception as e:
        print(f"An unexpected error occured:{e}")
        return None
    else:
        return result
    finally:
        print("Division operation complete")

print(divide(10,2))
print(divide(10,0))
print(divide("10",0))

#Docstrings
def add(a,b):
    """
    This function adds two numbers together.

    Args:
        a(int): The first number to be added
        b(int): The second number to be added

    Returns:
        int:The sum of the two input numbers.

    Examples:
        >>>add(3,4)
        7
    """
    return a+b

help(add) #display docsting
        
```
